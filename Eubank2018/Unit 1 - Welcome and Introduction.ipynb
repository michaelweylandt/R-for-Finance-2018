### Not covered today (but good to know!) 

- General Optimization: [`ROI`](https://cran.r-project.org/package=ROI), [`DEoptim`](https://cran.r-project.org/package=DEoptim), [`CVXR`](https://cran.r-project.org/package=CVXR)

- Factor Analysis
- Backtesting with [`quantstrat`](https://github.com/braverock/quantstrat)
- Analysis of high-frequency data: [`highfrequency`](https://cran.r-project.org/package=highfrequency)
- Econometric Analysis: [Econometrics in `R`](https://cran.r-project.org/web/views/Econometrics.html)
- Forecasting Time Series: [`forecast`](https://cran.r-project.org/package=forecast) & [`prophet`](https://cran.r-project.org/package=prophet)
- Multivariate Volatility:  [`rmgarch`](https://cran.r-project.org/package=rmgarch) & [`factorstochvol`](https://cran.r-project.org/package=factostochvol)
