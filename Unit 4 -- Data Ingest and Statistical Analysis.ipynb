## Linear models and friends: **lm**

Many functions in `R` allow use of a "formula" to specify a statistical model. 

`lm(y ~ x1 + x2, data=d)`

```r
library(nycflights13)
## How does distance impact flight time
model <- lm(air_time ~ distance, data=flights)
model
```

## Forecasting and Time Series Analysis

Common task -- forecasting a time series

`R` has great forecasting support thanks to the `forecast` library

Let's get started using some Australian wool production data

```r
library(forecast)
data(woolyrnq)
woolyrnq
```

```r
library(ggplot2)
autoplot(woolyrnq)
```

```r
tsdisplay(woolyrnq)
```

The `forecast` package provides a number of models to forecast the series: 

- Classical `ARIMA` -- `auto.arima`
- Seasonal + Trend decomposition -- `stlf`
- Neural networks (these are simple compared to state-of-the-art) -- `nnetar`
- Exponential smoothing -- `ets`

Let's compare them all

```r
m1 <- auto.arima(woolyrnq)
m1
```

```r
autoplot(forecast(m1))
```

```r
m2 <- stlf(woolyrnq)
print(m2)
```

```r
autoplot(forecast(m2))
```

```r
m3 <- nnetar(woolyrnq)
m3
```

```r
autoplot(forecast(m3))
```

```r
m4 <- ets(woolyrnq)
m4
```

```r
autoplot(forecast(m4))
```

The `forecast` package also provides some data on UK electricity demand -- `taylor`

```r
?taylor
```

```r
## The stlf decomposition works really well on this data
m5 <- stlf(taylor)
autoplot(forecast(m5))
```
